```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70,80]
y = [21,41,24,51,64,98,54,32]
```

```python
plt.plot(x,y,marker = "*", ms = 9, mec= "red", mfc = "green", linestyle = "dotted")
plt.grid()
plt.xlabel("Months")
plt.ylabel("Profit")
plt.title("Revenue")
plt.annotate(xy = (20,90), text = "hi")

for i in range(len(x)):
    plt.annotate(xy = (x[i], y[i]), text = f"({x[i]}, {y[i]}")
```
