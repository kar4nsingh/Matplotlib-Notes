```python
import seaborn as sns
```

```python
d = sns.load_dataset("anagrams")
d = d.drop(columns = "attnr")
d
```

```python
# Heatmap
sns.heatmap(d,vmin = 0, vmax = 8, cmap = "cividis", cbar = True)
```

```python
# boxplot
dd = sns.load_dataset("tips")
# dd
sns.boxplot(x = "day", y = "total_bill", data = dd, hue = "time", color = "black" )
```

```python
# pairplot
# dd
sns.pairplot(dd, hue = "day")
```

```python
# 3d

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
a = sns.load_dataset("penguins")
fig = plt.figure(figsize = (10,20))
ax = fig.add_subplot(111, projection = "3d")
ax.scatter(a["bill_length_mm"], a["bill_depth_mm"], a["flipper_length_mm"])
```
