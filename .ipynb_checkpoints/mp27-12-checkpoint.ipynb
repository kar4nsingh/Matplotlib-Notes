```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
c = ["red", "yellow", "blue", "green", "pink"]
label = [1,2,3,4,5]

plt.bar(x,y,color = c, label = label)
plt.title("Bar Chart",fontsize = 20,color = "red")
plt.xlabel("1-5",fontsize = 10,color = "blue")
plt.ylabel("10-50",fontsize = 15,color = "green")
plt.legend(loc = 2)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,20,15,40,50]
c = ["red", "yellow", "blue", "green", "pink"]
size = [100,200,300,40,500]

plt.scatter(x,y,color = c,s = size, marker= "*")
plt.show()
```
